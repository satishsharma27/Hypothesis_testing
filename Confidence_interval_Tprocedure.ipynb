```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:/All project/Titanic-Dataset.csv")
df.head(10)
```

```python
df["Fare"].isnull().sum()
```

```python
df["Fare"].plot(kind="kde")
```

```python
# sample size 30

samples = []

std = []

for i in range(10):
    x= df["Fare"].dropna().sample(30).values
    std.append(x.std())
    samples.append(x.tolist())
```

```python
samples
```

```python
std
```

```python
samples = np.array(samples)
samples
```

```python
sampling_means = samples.mean(axis=1)
```

```python
sampling_means
```

```python
sample_std = np.mean(std)
sample_std
```

```python
lower_limit = sampling_means.mean()-2.042*(sample_std/np.sqrt(30))
upper_limit = sampling_means.mean()+2.042*(sample_std/np.sqrt(30))
```

```python
print("The range","[",round(lower_limit,2),"-",round(upper_limit,2),"]")
```
